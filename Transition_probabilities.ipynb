```python
df1['Week_day']='Mon'
df2['Week_day']='Tue'
df3['Week_day']='Wed'
df4['Week_day']='Thurs'
df5['Week_day']='Friday'
```

```python
df = pd.concat([df1, df2, df3, df4, df5])
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
df.head()
```

```python
def time_stamps(df):
    df['hour'] = df.timestamp.dt.hour
    df['day'] = df.timestamp.dt.day
    df['month'] = df.timestamp.dt.month
    df['year'] = df.timestamp.dt.year
    return df
df=time_stamps(df)
```

```python
df['Customer_day']=df["customer_no"].astype(str) +' '+df["Week_day"]
```

```python
df=df.set_index('timestamp')
```

```python
##  In order to resample you need to group by Customer_day 

df=df.groupby('Customer_day').resample('T').ffill()
```
